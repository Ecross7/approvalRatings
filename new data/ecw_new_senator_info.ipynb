# 117th congressional

source: https://www.govtrack.us/congress/members

```python
df_billsIntro= pd.read_csv('/content/govtrack-stats-2022-senate-bills-introduced.csv')
df_companion = pd.read_csv('/content/govtrack-stats-2022-senate-bills-with-companion.csv')
df_cosponsored = pd.read_csv('/content/govtrack-stats-2022-senate-cosponsored.csv')
df_didCosponsor = pd.read_csv('/content/govtrack-stats-2022-senate-cosponsors.csv')
df_ideology = pd.read_csv('/content/govtrack-stats-2022-senate-ideology.csv')
df_leadership = pd.read_csv('/content/govtrack-stats-2022-senate-leadership.csv')
df_missed = pd.read_csv('/content/govtrack-stats-2022-senate-missed-votes.csv')
```

```python
df_billsIntro.head()
```

```python
df_billsIntro = df_billsIntro[['bills-introduced','id','bioguide_id','name']]
df_billsIntro.head()
```

```python
df_companion = df_companion[['bills-with-companion','id','bioguide_id','name']]
df_companion.head()
```

```python
df_cosponsored =df_cosponsored[['cosponsored','id','bioguide_id','name']]
df_cosponsored.head()
```

```python
df_didCosponsor =df_didCosponsor[['cosponsors','id','bioguide_id','name']]
df_didCosponsor.head()
```

```python
df_ideology =df_ideology[['ideology','id','bioguide_id']]
df_ideology.head()
```

```python
df_leadership = df_leadership[['leadership','id','bioguide_id','name']]
df_leadership.head()
```

```python
df_missed = df_missed[['missed-votes','id','bioguide_id','name']]
df_missed.head()
```

```python
df_ideology
```

```python
frames = [df_billsIntro, df_companion,df_cosponsored, df_didCosponsor,df_ideology, df_leadership, df_missed]
```

```python
merged = pd.concat(frames, join='inner',axis=1)
```

```python
merged = merged.loc[:,~merged.columns.duplicated()].copy()
```

```python
merged['congressional'] = 117
```

```python
merged.head()
```

```python
merged.to_csv('117_out.csv')
```
