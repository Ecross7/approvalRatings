# add congressional number

```python
from datetime import datetime
```

```python
def dateInRange(date,startDate,endDate):
  return startDate <= date <= endDate

def datetoCongress(date):
  '''
  year = 0
  t = date.to_pydatetime()
  #print(t.day)
  if t.month in [11,12]:
    year = int(t.year) + 1
    date = datetime(year,t.month,t.day)
  '''

  if dateInRange(date, datetime(2023,1,1),datetime(2025,1,1)):
    return 118
  elif dateInRange(date, datetime(2021,1,1),datetime(2023,1,1)):
    return 117
  elif dateInRange(date, datetime(2019,1,1),datetime(2021,1,1)):
    return 116
  elif dateInRange(date, datetime(2017,1,1),datetime(2019,1,1)):
    return 115
  else:
    return 114

def getCycle(num):
  num = int(num)
  if num == 118:
    return 2024
  elif num == 117:
    return 2022
  elif num == 116:
    return 2020
  elif num == 115:
    return 2018
  else:
    return 2016
```

```python
df.columns
```

```python
#add congressional number at the time the poll was taken
df['congressional'] = df['end_date'].apply(lambda x: datetoCongress(x))

#df.drop(columns='congressional',inplace=True)
```
