# Merge both stages of images to one folder

```python
import os
import shutil

# Source directory containing images of senators
source_dir = "sen2_images"

# Destination directory to merge images
merged_dir = "merged_sen2_images"

# Create the merged directory if it doesn't exist
os.makedirs(merged_dir, exist_ok=True)

# Iterate over each senator directory
for senator_dir in os.listdir(source_dir):
    senator_path = os.path.join(source_dir, senator_dir)
    # Skip any non-directory files
    if not os.path.isdir(senator_path):
        continue
    # Iterate over each image file in the senator directory
    for filename in os.listdir(senator_path):
        image_path = os.path.join(senator_path, filename)
        # Copy the image file to the merged directory
        shutil.copy(image_path, merged_dir)

print("Images merged successfully!")
```

# Emotion recogntion and Image analysis using Deep Face [2nd stage ]

```python
from deepface import DeepFace
import pandas as pd
import os

# Define weight sets for valence calculation
weights_set1 = {
    "happy": 0.4,
    "surprise": 0.3,
    "angry": -0.2,
    "sad": -0.1,
    "fear": -0.1,
    "disgust": -0.1,
    "neutral": 0.0
}

weights_set2 = {
    "happy": 0.3,
    "surprise": 0.2,
    "angry": -0.3,
    "sad": -0.2,
    "fear": -0.2,
    "disgust": -0.1,
    "neutral": 0.0
}

weights_set3 = {
    "happy": 0.5,
    "surprise": 0.2,
    "angry": -0.1,
    "sad": -0.3,
    "fear": -0.1,
    "disgust": -0.1,
    "neutral": 0.0
}

# Initialize an empty DataFrame to store the analysis results
face_df2 = pd.DataFrame(columns=["Senator", "Age", "Gender", "Emotion", "Race", "Val1", "Val2", "Val3"])

# Define the folder path containing the images
img_folder_path = '/Users/mark/Final Project/Final_Images_Set/'
img_files = os.listdir(img_folder_path)

# Iterate through each image in the folder
for img_file in img_files:
    try:
        # Extract senator name from the image filename
        senator_name = img_file.split('_')[0]

        # Analyze image using DeepFace
        result = DeepFace.analyze(img_path=os.path.join(img_folder_path, img_file))

        # Calculate valences using different weight sets
        val1 = sum(int(result[0]["emotion"][emotion]) * weight for emotion, weight in weights_set1.items()) / 100
        val2 = sum(int(result[0]["emotion"][emotion]) * weight for emotion, weight in weights_set2.items()) / 100
        val3 = sum(int(result[0]["emotion"][emotion]) * weight for emotion, weight in weights_set3.items()) / 100

        # Append analysis results to the DataFrame
        face_df2 = face_df2.append({
            "Senator": senator_name,
            "Age": result[0]["age"],
            "Gender": result[0]["dominant_gender"],
            "Emotion": result[0]["dominant_emotion"],
            "Race": result[0]["dominant_race"],
            "Val1": val1,
            "Val2": val2,
            "Val3": val3
        }, ignore_index=True)
    except ValueError as e:
        print(f"Error analyzing image '{img_file}': {e}")
        continue
```
